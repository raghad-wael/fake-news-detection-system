```python
# fake_news_combined_full_corrected.py
# Full pipeline to merge 3 pre-cleaned datasets (ISOT, TextDB3, WELFake)
# - Loads cleaned CSVs from /kaggle/input/cleaned-dataset/
# - Performs final cross-dataset deduplication
# - Runs complete training pipeline (ML + Deep Learning models)
# - Evaluates on concatenated Title + Article Text

import os
import json
import re
import pickle
from datetime import datetime
from io import BytesIO
import numpy as np
import pandas as pd
from tqdm.auto import tqdm

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.metrics import roc_curve, auc
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.utils import shuffle

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from transformers import (BertTokenizer, BertForSequenceClassification,
                          RobertaTokenizer, RobertaForSequenceClassification,
                          DistilBertTokenizer, DistilBertForSequenceClassification,
                          get_linear_schedule_with_warmup)
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# =============================================================================
# CONFIGURATION
# =============================================================================
CLEANED_ISOT_PATH = "/kaggle/input/datasets/raghadwael21/cleaned-dataset/cleaned_ISOT_Dataset.csv"
CLEANED_TEXTDB3_PATH = "/kaggle/input/datasets/raghadwael21/cleaned-dataset/cleaned_TextDB3_Dataset.csv"
CLEANED_WELFAKE_PATH = "/kaggle/input/datasets/raghadwael21/cleaned-dataset/cleaned_WELFake_Dataset.csv"
OUTPUT_COMBINED_PATH = "/kaggle/working/combined_dataset.csv" 
CHECKPOINT_DIR = "checkpoints"

# Offline transformer paths (if available)
bert_path = "/kaggle/input/datasets/raghadwael21/hugging-face-downloaded/bert-base-uncased/bert-base-uncased"
roberta_path = "/kaggle/input/datasets/raghadwael21/hugging-face-downloaded/roberta-base/roberta-base"
distilbert_path = "/kaggle/input/datasets/raghadwael21/hugging-face-downloaded/distilbert-base-uncased/distilbert-base-uncased"

# Model flags
USE_TRANSFORMERS = True
USE_LSTM = True

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🔧 Device: {device}")

if torch.cuda.is_available():
    print(f"    GPU: {torch.cuda.get_device_name(0)}")
    print(f"    Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")

os.makedirs(CHECKPOINT_DIR, exist_ok=True)


# =============================================================================
# UTILITIES
# =============================================================================
def clean_text(text):
    """Basic text cleaning"""
    if pd.isna(text):
        return ""
    txt = str(text).lower()
    txt = re.sub(r'http\S+|www\S+|https\S+', ' ', txt, flags=re.MULTILINE)
    txt = re.sub(r'\S+@\S+', ' ', txt)
    txt = re.sub(r'<.*?>', ' ', txt)
    txt = re.sub(r'@\w+|#\w+', ' ', txt)
    txt = re.sub(r'[^a-zA-Z\s]', ' ', txt)
    txt = re.sub(r'\s+', ' ', txt).strip()
    return txt

def safe_json_load(x):
    """Safely load JSON from string or dict"""
    if pd.isna(x):
        return {}
    if isinstance(x, dict):
        return x
    try:
        return json.loads(str(x).replace("'", '"'))
    except Exception:
        try:
            return eval(x)
        except Exception:
            return {}

def remove_near_duplicates(df, text_column='text', similarity_threshold=0.98, batch_size=500):
    """Remove near-duplicates using TF-IDF cosine similarity"""
    print(f"\n🔍 Removing near-duplicates (similarity >= {similarity_threshold})...")
    original_size = len(df)
    
    if original_size == 0:
        return df
    
    df = df.reset_index(drop=True)
    
    print("    Creating TF-IDF vectors...")
    vectorizer = TfidfVectorizer(max_features=2000, stop_words='english')
    
    try:
        tfidf_matrix = vectorizer.fit_transform(df[text_column].astype(str))
    except Exception as e:
        print(f"    ⚠️ TF-IDF failed: {e}")
        return df
    
    duplicates_to_remove = set()
    n_samples = tfidf_matrix.shape[0]
    
    print(f"    Checking {n_samples:,} samples...")
    for i in tqdm(range(0, n_samples, batch_size), desc="    Batches"):
        batch_end = min(i + batch_size, n_samples)
        similarities = cosine_similarity(tfidf_matrix[i:batch_end], tfidf_matrix)
        
        for local_idx, row_similarities in enumerate(similarities):
            global_idx = i + local_idx
            
            if global_idx in duplicates_to_remove:
                continue
            
            duplicate_indices = np.where(row_similarities >= similarity_threshold)[0]
            duplicate_indices = duplicate_indices[duplicate_indices > global_idx]
            duplicates_to_remove.update(duplicate_indices.tolist())
    
    indices_to_keep = [i for i in range(n_samples) if i not in duplicates_to_remove]
    df_cleaned = df.iloc[indices_to_keep].reset_index(drop=True)
    
    removed = original_size - len(df_cleaned)
    print(f"    ✓ Removed {removed:,} near-duplicates ({removed/original_size*100:.2f}%)")
    print(f"    ✓ Retained {len(df_cleaned):,} unique samples")
    
    return df_cleaned


# =============================================================================
# DATASET LOADING AND MERGING
# =============================================================================
def load_and_merge_cleaned_datasets():
    """Load all 3 pre-cleaned datasets and merge them"""
    print("\n" + "="*100)
    print("📚 LOADING PRE-CLEANED DATASETS")
    print("="*100)
    
    datasets_info = [
        (CLEANED_ISOT_PATH, "ISOT"),
        (CLEANED_TEXTDB3_PATH, "TextDB3"),
        (CLEANED_WELFAKE_PATH, "WELFake")
    ]
    
    dfs = []
    
    for path, source_name in datasets_info:
        if not os.path.exists(path):
            print(f"⚠️  {source_name}: File not found at {path} - SKIPPING")
            continue
        
        try:
            print(f"\n📂 Loading {source_name}...")
            df = pd.read_csv(path)
            
            # Drop Unnamed columns (from index)
            unnamed_cols = [col for col in df.columns if 'Unnamed' in str(col)]
            if unnamed_cols:
                df = df.drop(unnamed_cols, axis=1)
            
            # Check for required columns
            if 'text' not in df.columns or 'label' not in df.columns:
                print(f"    ❌ Missing 'text' or 'label' column - SKIPPING")
                continue
            
            # Ensure label is integer
            df['label'] = df['label'].astype(int)
            
            # Add optional columns if missing
            optional_cols = ['title', 'subject', 'date', 'fetch_metadata', 'image_features']
            for col in optional_cols:
                if col not in df.columns:
                    df[col] = None
            
            # Parse fetch_metadata
            if 'fetch_metadata' in df.columns:
                df['fetch_metadata'] = df['fetch_metadata'].apply(safe_json_load)
            
            # Parse image_features
            def parse_img(x):
                if pd.isna(x):
                    return None
                if isinstance(x, list):
                    return x
                try:
                    return json.loads(str(x).replace("'", '"'))
                except Exception:
                    try:
                        return eval(x)
                    except Exception:
                        return None
            
            if 'image_features' in df.columns:
                df['image_features'] = df['image_features'].apply(parse_img)
            
            # Add source column
            df['source'] = source_name
            
            # Select columns
            keep_cols = ['title', 'text', 'subject', 'date', 'label', 'fetch_metadata', 'image_features', 'source']
            df = df[keep_cols]
            
            dfs.append(df)
            print(f"    ✓ Loaded {len(df):,} samples from {source_name}")
            
        except Exception as e:
            print(f"    ❌ Error loading {source_name}: {e}")
            import traceback
            traceback.print_exc()
            continue
    
    if not dfs:
        raise ValueError("❌ No datasets were successfully loaded!")
    
    # Merge all datasets
    print("\n🔗 Merging all datasets...")
    combined = pd.concat(dfs, ignore_index=True)
    original_size = len(combined)
    
    # Final cleaning on merged dataset
    print("\n🧹 Final cleaning on merged dataset...")
    
    combined = combined.dropna(subset=['text', 'label']).reset_index(drop=True)
    combined['label'] = combined['label'].astype(int)
    
    print("    Combining Title and Text content...")
    
    # -----------------------------------------------------------------------------
    # [MODIFIED: Concatenate Title and Text before cleaning]
    # -----------------------------------------------------------------------------
    combined['title'] = combined['title'].fillna("")
    combined['text'] = combined['title'] + " - " + combined['text']
    # -----------------------------------------------------------------------------

    combined['text'] = combined['text'].apply(clean_text)
    
    combined = combined[combined['text'].str.len() >= 10].reset_index(drop=True)
    combined = combined.drop_duplicates(subset=['text']).reset_index(drop=True)
    
    # Remove near-duplicates across merged dataset
    combined = remove_near_duplicates(combined, text_column='text', similarity_threshold=0.98, batch_size=1000)
    
    # Shuffle
    print("\n🔀 Shuffling combined dataset...")
    combined = shuffle(combined, random_state=42).reset_index(drop=True)
    
    # Save combined dataset
    combined.to_csv(OUTPUT_COMBINED_PATH, index=False)
    print(f"\n💾 Saved combined dataset to: {OUTPUT_COMBINED_PATH}")
    
    return combined


# =============================================================================
# MODEL CLASSES
# =============================================================================
class NewsDataset(Dataset):
    """Custom Dataset for Transformer models"""
    def __init__(self, texts, labels, tokenizer, max_len=256):
        self.texts = [str(t) for t in (texts.values if hasattr(texts, 'values') else texts)]
        self.labels = list(labels.values if hasattr(labels, 'values') else labels)
        self.tokenizer = tokenizer
        self.max_len = max_len
    
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        text = self.texts[idx][:1000]
        label = int(self.labels[idx])
        
        encoding = self.tokenizer(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )
        
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'label': torch.tensor(label, dtype=torch.long)
        }


class TransformerClassifier:
    """Unified transformer classifier (BERT/RoBERTa/DistilBERT)"""
    def __init__(self, model_type='bert', model_name=None, max_len=256, batch_size=16):
        self.model_type = model_type
        self.max_len = max_len
        self.batch_size = batch_size
        self.history = {'train_loss': [], 'val_loss': [], 'val_acc': []}
        
        model_names = {
            'bert': 'bert-base-uncased',
            'roberta': 'roberta-base',
            'distilbert': 'distilbert-base-uncased'
        }
        
        if model_name is None:
            model_name = model_names.get(model_type, 'bert-base-uncased')
        
        self.model_name = model_name
        
        local_paths_map = {
            'bert-base-uncased': bert_path,
            'roberta-base': roberta_path,
            'distilbert-base-uncased': distilbert_path
        }
        
        local_path = local_paths_map.get(model_name)
        self.local_only = bool(local_path and os.path.exists(local_path))
        self.source = local_path if self.local_only else model_name
        
        if model_type == 'bert':
            self.tokenizer = BertTokenizer.from_pretrained(self.source, local_files_only=self.local_only)
            self.model_class = BertForSequenceClassification
        elif model_type == 'roberta':
            self.tokenizer = RobertaTokenizer.from_pretrained(self.source, local_files_only=self.local_only)
            self.model_class = RobertaForSequenceClassification
        elif model_type == 'distilbert':
            self.tokenizer = DistilBertTokenizer.from_pretrained(self.source, local_files_only=self.local_only)
            self.model_class = DistilBertForSequenceClassification
        
        self.model = None
    
    def train(self, X_train, y_train, X_val, y_val, epochs=2, lr=2e-5):
        """Train the model"""
        print(f"\n🚀 Training {self.model_type.upper()}...")
        
        train_dataset = NewsDataset(X_train, y_train, self.tokenizer, self.max_len)
        val_dataset = NewsDataset(X_val, y_val, self.tokenizer, self.max_len)
        
        train_loader = DataLoader(train_dataset, batch_size=self.batch_size, shuffle=True, num_workers=0)
        val_loader = DataLoader(val_dataset, batch_size=self.batch_size, num_workers=0)
        
        self.model = self.model_class.from_pretrained(
            self.source,
            num_labels=2,
            output_attentions=False,
            output_hidden_states=False,
            local_files_only=self.local_only
        )
        self.model.to(device)
        
        optimizer = AdamW(self.model.parameters(), lr=lr, eps=1e-8)
        total_steps = len(train_loader) * epochs
        scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=total_steps)
        
        best_val_acc = 0
        for epoch in range(epochs):
            print(f"\n📊 Epoch {epoch + 1}/{epochs}")
            self.model.train()
            total_loss = 0
            
            progress_bar = tqdm(train_loader, desc='Training')
            for batch in progress_bar:
                optimizer.zero_grad()
                
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels = batch['label'].to(device)
                
                outputs = self.model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
                loss = outputs.loss
                total_loss += loss.item()
                
                loss.backward()
                torch.nn.utils.clip_grad_norm_(self.model.parameters(), 1.0)
                optimizer.step()
                scheduler.step()
                
                progress_bar.set_postfix({'loss': f'{loss.item():.4f}'})
            
            avg_train_loss = total_loss / len(train_loader)
            val_loss, val_accuracy = self._evaluate(val_loader)
            
            self.history['train_loss'].append(avg_train_loss)
            self.history['val_loss'].append(val_loss)
            self.history['val_acc'].append(val_accuracy)
            
            print(f"    ✓ Train Loss: {avg_train_loss:.4f} | Val Loss: {val_loss:.4f} | Val Accuracy: {val_accuracy:.4f}")
            
            if val_accuracy > best_val_acc:
                best_val_acc = val_accuracy
        
        torch.cuda.empty_cache()
        return self.model
    
    def _evaluate(self, data_loader):
        """Evaluate model"""
        self.model.eval()
        total_loss = 0
        correct = 0
        total = 0
        
        with torch.no_grad():
            for batch in data_loader:
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels = batch['label'].to(device)
                
                outputs = self.model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
                total_loss += outputs.loss.item()
                
                _, predicted = torch.max(outputs.logits, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()
        
        avg_val_loss = total_loss / len(data_loader) if len(data_loader) > 0 else 0.0
        accuracy = correct / total if total > 0 else 0.0
        return avg_val_loss, accuracy
    
    def predict(self, texts):
        """Predict labels"""
        self.model.eval()
        predictions = []
        probabilities = []
        
        dataset = NewsDataset(texts, [0] * len(texts), self.tokenizer, self.max_len)
        data_loader = DataLoader(dataset, batch_size=self.batch_size, num_workers=0)
        
        with torch.no_grad():
            for batch in tqdm(data_loader, desc='Predicting', leave=False):
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                
                outputs = self.model(input_ids=input_ids, attention_mask=attention_mask)
                probs = torch.nn.functional.softmax(outputs.logits, dim=1)
                _, preds = torch.max(outputs.logits, 1)
                
                predictions.extend(preds.cpu().numpy())
                probabilities.extend(probs.cpu().numpy())
        
        return np.array(predictions), np.array(probabilities)


class LSTMClassifier(nn.Module):
    """Bidirectional LSTM classifier"""
    def __init__(self, vocab_size, embedding_dim=128, hidden_dim=128, num_layers=2, dropout=0.3):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers, 
                            batch_first=True, dropout=dropout, bidirectional=True)
        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Sequential(
            nn.Linear(hidden_dim * 2, 64),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(64, 2)
        )
        
    def forward(self, x, lengths):
        embedded = self.embedding(x)
        packed = nn.utils.rnn.pack_padded_sequence(embedded, lengths.cpu(), 
                                                   batch_first=True, enforce_sorted=False)
        lstm_out, (hidden, cell) = self.lstm(packed)
        hidden = torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim=1)
        hidden = self.dropout(hidden)
        out = self.fc(hidden)
        return out


class LSTMWrapper:
    """LSTM wrapper"""
    def __init__(self, vocab_size=5000, max_len=256, embedding_dim=128, 
                 hidden_dim=128, num_layers=2, batch_size=64, epochs=3):
        self.vocab_size = vocab_size
        self.max_len = max_len
        self.embedding_dim = embedding_dim
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        self.batch_size = batch_size
        self.epochs = epochs
        self.model = None
        self.word_to_idx = {}
        self.history = {'train_loss': [], 'val_loss': [], 'val_acc': []}
        
    def _build_vocab(self, texts):
        word_freq = {}
        texts_list = texts.values if hasattr(texts, 'values') else texts
        for text in texts_list:
            for word in str(text).lower().split()[:200]:
                word_freq[word] = word_freq.get(word, 0) + 1
        
        sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
        self.word_to_idx = {word: idx + 1 for idx, (word, _) in enumerate(sorted_words[:self.vocab_size-1])}
        self.word_to_idx['<PAD>'] = 0
        
    def _texts_to_sequences(self, texts):
        texts_list = texts.values if hasattr(texts, 'values') else texts
        sequences = []
        lengths = []
        for text in texts_list:
            words = str(text).lower().split()[:self.max_len]
            seq = [self.word_to_idx.get(word, 0) for word in words]
            lengths.append(min(len(seq), self.max_len))
            seq = seq + [0] * (self.max_len - len(seq))
            sequences.append(seq[:self.max_len])
        return torch.LongTensor(sequences), torch.LongTensor(lengths)
    
    def train(self, X_train, y_train, X_val, y_val):
        """Train LSTM"""
        print("\n🚀 Training LSTM...")
        self._build_vocab(X_train)
        
        X_train_seq, train_lengths = self._texts_to_sequences(X_train)
        X_val_seq, val_lengths = self._texts_to_sequences(X_val)
        
        y_train_tensor = torch.LongTensor(y_train.values if hasattr(y_train, 'values') else y_train)
        y_val_tensor = torch.LongTensor(y_val.values if hasattr(y_val, 'values') else y_val)
        
        train_dataset = torch.utils.data.TensorDataset(X_train_seq, train_lengths, y_train_tensor)
        val_dataset = torch.utils.data.TensorDataset(X_val_seq, val_lengths, y_val_tensor)
        
        train_loader = DataLoader(train_dataset, batch_size=self.batch_size, shuffle=True, num_workers=0)
        val_loader = DataLoader(val_dataset, batch_size=self.batch_size, num_workers=0)
        
        self.model = LSTMClassifier(len(self.word_to_idx), self.embedding_dim, 
                                       self.hidden_dim, self.num_layers)
        self.model.to(device)
        
        criterion = nn.CrossEntropyLoss()
        optimizer = torch.optim.Adam(self.model.parameters(), lr=0.001)
        
        for epoch in range(self.epochs):
            self.model.train()
            total_loss = 0
            
            progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{self.epochs}')
            for sequences, lengths, labels in progress_bar:
                sequences, lengths, labels = sequences.to(device), lengths.to(device), labels.to(device)
                
                optimizer.zero_grad()
                outputs = self.model(sequences, lengths)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                
                total_loss += loss.item()
                progress_bar.set_postfix({'loss': f'{loss.item():.4f}'})
            
            avg_train_loss = total_loss / len(train_loader)
            
            # Validation
            self.model.eval()
            total_val_loss = 0
            correct = 0
            total = 0
            with torch.no_grad():
                for sequences, lengths, labels in val_loader:
                    sequences, lengths, labels = sequences.to(device), lengths.to(device), labels.to(device)
                    outputs = self.model(sequences, lengths)
                    
                    batch_loss = criterion(outputs, labels)
                    total_val_loss += batch_loss.item()
                    
                    _, predicted = torch.max(outputs, 1)
                    total += labels.size(0)
                    correct += (predicted == labels).sum().item()
            
            avg_val_loss = total_val_loss / len(val_loader) if len(val_loader) > 0 else 0.0
            val_accuracy = correct / total if total > 0 else 0
            
            self.history['train_loss'].append(avg_train_loss)
            self.history['val_loss'].append(avg_val_loss)
            self.history['val_acc'].append(val_accuracy)
            
            print(f"    ✓ Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f} | Val Acc: {val_accuracy:.4f}")
        
        torch.cuda.empty_cache()
        return self.model
    
    def predict(self, texts):
        self.model.eval()
        X_seq, lengths = self._texts_to_sequences(texts)
        dataset = torch.utils.data.TensorDataset(X_seq, lengths)
        data_loader = DataLoader(dataset, batch_size=self.batch_size, num_workers=0)
        
        predictions = []
        probabilities = []
        
        with torch.no_grad():
            for sequences, lengths in data_loader:
                sequences, lengths = sequences.to(device), lengths.to(device)
                outputs = self.model(sequences, lengths)
                probs = torch.nn.functional.softmax(outputs, dim=1)
                _, preds = torch.max(outputs, 1)
                
                predictions.extend(preds.cpu().numpy())
                probabilities.extend(probs.cpu().numpy())
        
        return np.array(predictions), np.array(probabilities)


# =============================================================================
# TRAINING AND EVALUATION
# =============================================================================
def train_and_evaluate(df, models_dict):
    """Train and evaluate all models"""
    
    X = df['text']
    y = df['label'].astype(int)
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.10, random_state=42, stratify=y_train)
    
    print(f"\n📊 Data Split:")
    print(f"    Train: {len(X_train):,} samples")
    print(f"    Val:   {len(X_val):,} samples")
    print(f"    Test:  {len(X_test):,} samples")
    print("="*80)
    
    results = {}
    trained_models = {}
    
    for model_name, model_info in models_dict.items():
        print(f"\n{'='*80}")
        print(f"🎯 Training: {model_name}")
        print(f"{'='*80}")
        
        try:
            if model_info['type'] == 'sklearn':
                vectorizer, model = model_info['vectorizer'], model_info['model']
                print("    Vectorizing text...")
                X_train_vec = vectorizer.fit_transform(X_train)
                X_test_vec = vectorizer.transform(X_test)
                
                print("    Training model...")
                model.fit(X_train_vec, y_train)
                y_pred = model.predict(X_test_vec)
                y_pred_proba = model.predict_proba(X_test_vec) if hasattr(model, 'predict_proba') else None
                trained_models[model_name] = {'model': model, 'vectorizer': vectorizer, 'type': 'sklearn'}
                
            elif model_info['type'] == 'transformer':
                mdl = model_info['model']
                mdl.train(X_train, y_train, X_val, y_val)
                y_pred, y_pred_proba = mdl.predict(X_test)
                trained_models[model_name] = {'model': mdl, 'type': 'transformer'}
                
            elif model_info['type'] == 'lstm':
                mdl = model_info['model']
                mdl.train(X_train, y_train, X_val, y_val)
                y_pred, y_pred_proba = mdl.predict(X_test)
                trained_models[model_name] = {'model': mdl, 'type': 'lstm'}
            
            accuracy = accuracy_score(y_test, y_pred)
            precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
            recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
            f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
            
            results[model_name] = {
                'accuracy': accuracy,
                'precision': precision,
                'recall': recall,
                'f1': f1,
                'y_pred': y_pred,
                'y_pred_proba': y_pred_proba,
                'y_test': y_test
            }
            
            print(f"\n✅ Results:")
            print(f"    Accuracy:  {accuracy:.4f}")
            print(f"    Precision: {precision:.4f}")
            print(f"    Recall:    {recall:.4f}")
            print(f"    F1 Score:  {f1:.4f}")
            
            checkpoint_file = os.path.join(CHECKPOINT_DIR, f"checkpoint_{model_name}_{datetime.now().strftime('%Y%m%d_%H%M%S')}.pkl")
            with open(checkpoint_file, 'wb') as f:
                pickle.dump({'model_name': model_name, 'results': results[model_name]}, f)
                
            if model_info['type'] == 'transformer':
                model_path = os.path.join(CHECKPOINT_DIR, f"{model_name}_final_weights")
                mdl.model.save_pretrained(model_path)
                mdl.tokenizer.save_pretrained(model_path)
                print(f"    💾 Saved transformer weights to {model_path}")
            elif model_info['type'] == 'lstm':
                model_path = os.path.join(CHECKPOINT_DIR, f"{model_name}_final_weights.pth")
                torch.save(mdl.model.state_dict(), model_path)
                print(f"    💾 Saved LSTM weights to {model_path}")
            
        except Exception as e:
            print(f"❌ Error training {model_name}: {str(e)}")
            results[model_name] = None
    
    print(f"\n{'='*80}")
    print("🤝 Creating Ensemble Models")
    print(f"{'='*80}")
    
    predictions_dict = {}
    probabilities_dict = {}
    
    for model_name, model_data in trained_models.items():
        try:
            if model_data['type'] == 'sklearn':
                vectorizer, model = model_data['vectorizer'], model_data['model']
                X_test_vec = vectorizer.transform(X_test)
                if hasattr(model, 'predict'):
                    predictions_dict[model_name] = model.predict(X_test_vec)
                if hasattr(model, 'predict_proba'):
                    probabilities_dict[model_name] = model.predict_proba(X_test_vec)
            elif model_data['type'] in ['transformer', 'lstm']:
                model = model_data['model']
                preds, proba = model.predict(X_test)
                predictions_dict[model_name] = preds
                probabilities_dict[model_name] = proba
        except Exception as e:
            continue
    
    ensemble_results = {}
    
    if probabilities_dict:
        print("\n1️⃣  Soft Voting Ensemble")
        avg_proba = np.mean(list(probabilities_dict.values()), axis=0)
        soft_pred = np.argmax(avg_proba, axis=1)
        
        ensemble_results['Soft Voting'] = {
            'accuracy': accuracy_score(y_test, soft_pred),
            'precision': precision_score(y_test, soft_pred, average='weighted', zero_division=0),
            'recall': recall_score(y_test, soft_pred, average='weighted', zero_division=0),
            'f1': f1_score(y_test, soft_pred, average='weighted', zero_division=0)
        }
        print(f"    F1 Score: {ensemble_results['Soft Voting']['f1']:.4f}")
    
    if predictions_dict:
        print("\n2️⃣  Hard Voting Ensemble")
        votes = np.vstack(list(predictions_dict.values()))
        hard_pred = np.apply_along_axis(lambda x: np.bincount(x.astype(int)).argmax(), axis=0, arr=votes)
        
        ensemble_results['Hard Voting'] = {
            'accuracy': accuracy_score(y_test, hard_pred),
            'precision': precision_score(y_test, hard_pred, average='weighted', zero_division=0),
            'recall': recall_score(y_test, hard_pred, average='weighted', zero_division=0),
            'f1': f1_score(y_test, hard_pred, average='weighted', zero_division=0)
        }
        print(f"    F1 Score: {ensemble_results['Hard Voting']['f1']:.4f}")
    
    return results, ensemble_results, trained_models


# =============================================================================
# VISUALIZATIONS
# =============================================================================
def visualize_results(results, ensemble_results, df):
    """Create comprehensive visualizations"""
    rows = []
    for model_name, metrics in results.items():
        if metrics:
            rows.append({
                'Model': model_name,
                'Type': 'Individual',
                'Accuracy': metrics['accuracy'],
                'Precision': metrics['precision'],
                'Recall': metrics['recall'],
                'F1 Score': metrics['f1']
            })
    
    for model_name, metrics in (ensemble_results or {}).items():
        rows.append({
            'Model': model_name,
            'Type': 'Ensemble',
            'Accuracy': metrics['accuracy'],
            'Precision': metrics['precision'],
            'Recall': metrics['recall'],
            'F1 Score': metrics['f1']
        })
    
    df_results = pd.DataFrame(rows).sort_values('F1 Score', ascending=False)
    df_results.to_csv("combined_model_results.csv", index=False)
    print("\n💾 Saved results to: combined_model_results.csv")
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    
    ax1 = axes[0, 0]
    colors = ['#3498db' if t == 'Individual' else '#e74c3c' for t in df_results['Type']]
    df_results.plot(x='Model', y='F1 Score', kind='bar', ax=ax1, color=colors, legend=False)
    ax1.set_title('F1 Score Comparison - Combined Dataset', fontsize=14, fontweight='bold')
    ax1.set_ylabel('F1 Score', fontsize=12)
    ax1.set_xlabel('')
    ax1.grid(axis='y', alpha=0.3)
    plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45, ha='right')
    
    ax2 = axes[0, 1]
    metrics_df = df_results.set_index('Model')[['Accuracy', 'Precision', 'Recall', 'F1 Score']]
    metrics_df.plot(kind='bar', ax=ax2, width=0.8)
    ax2.set_title('All Metrics Comparison', fontsize=14, fontweight='bold')
    ax2.set_ylabel('Score', fontsize=12)
    ax2.set_xlabel('')
    ax2.legend(loc='lower right')
    ax2.grid(axis='y', alpha=0.3)
    plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45, ha='right')
    
    ax3 = axes[1, 0]
    top_models = df_results.head(min(8, len(df_results)))
    colors_top = plt.cm.RdYlGn(np.linspace(0.5, 0.9, len(top_models)))
    ax3.barh(range(len(top_models)), top_models['F1 Score'], color=colors_top)
    ax3.set_yticks(range(len(top_models)))
    ax3.set_yticklabels(top_models['Model'])
    ax3.set_xlabel('F1 Score', fontsize=12)
    ax3.set_title('Top Models by F1 Score', fontsize=14, fontweight='bold')
    ax3.invert_yaxis()
    ax3.grid(axis='x', alpha=0.3)
    
    ax4 = axes[1, 1]
    source_counts = df['source'].value_counts()
    colors_pie = plt.cm.Set3(range(len(source_counts)))
    ax4.pie(source_counts.values, labels=source_counts.index, autopct='%1.1f%%',
            startangle=90, colors=colors_pie)
    ax4.set_title('Dataset Source Distribution', fontsize=14, fontweight='bold')
    
    plt.suptitle('Fake News Detection - Combined Dataset Results', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.savefig('combined_results.png', dpi=150, bbox_inches='tight')
    print("✅ Visualization saved as: combined_results.png")
    plt.show()
    
    return df_results

def create_confusion_matrix(results):
    """Create confusion matrix for best model"""
    valid_results = {k: v for k, v in results.items() if v}
    if not valid_results:
        return
    
    best_model_name = max(valid_results.items(), key=lambda x: x[1]['f1'])[0]
    best_metrics = valid_results[best_model_name]
    
    y_true = best_metrics['y_test']
    y_pred = best_metrics['y_pred']
    
    cm = confusion_matrix(y_true, y_pred)
    
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                annot_kws={'size': 16, 'weight': 'bold'})
    plt.ylabel('True Label', fontsize=12)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.title(f'Confusion Matrix - Best Model: {best_model_name}', fontsize=14, fontweight='bold')
    plt.xticks([0.5, 1.5], ['Real (0)', 'Fake (1)'])
    plt.yticks([0.5, 1.5], ['Real (0)', 'Fake (1)'])
    
    plt.tight_layout()
    plt.savefig('combined_confusion_matrix.png', dpi=150, bbox_inches='tight')
    print("✅ Confusion matrix saved as: combined_confusion_matrix.png")
    plt.show()

def plot_roc_curves(results):
    """Plot ROC curves for models that output probabilities"""
    plt.figure(figsize=(10, 8))
    
    has_curves = False
    for model_name, metrics in results.items():
        if metrics and metrics.get('y_pred_proba') is not None:
            # y_pred_proba usually has shape (n_samples, 2), we want probability of class 1 (Fake)
            if len(metrics['y_pred_proba'].shape) == 2:
                y_scores = metrics['y_pred_proba'][:, 1]
            else:
                y_scores = metrics['y_pred_proba']
                
            fpr, tpr, _ = roc_curve(metrics['y_test'], y_scores)
            roc_auc = auc(fpr, tpr)
            
            plt.plot(fpr, tpr, lw=2, label=f'{model_name} (AUC = {roc_auc:.3f})')
            has_curves = True

    if not has_curves:
        print("⚠️ No probability predictions found. Skipping ROC curve plot.")
        plt.close()
        return

    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=14, fontweight='bold')
    plt.legend(loc="lower right")
    plt.grid(alpha=0.3)
    plt.savefig('combined_roc_curves.png', dpi=150, bbox_inches='tight')
    print("✅ ROC Curves saved as: combined_roc_curves.png")
    plt.show()

def plot_training_histories(trained_models):
    """Plot training and validation loss curves for deep learning models"""
    dl_models = {k: v for k, v in trained_models.items() if v['type'] in ['transformer', 'lstm']}
    
    if not dl_models:
        print("⚠️ No deep learning models found to plot training histories.")
        return
        
    num_models = len(dl_models)
    fig, axes = plt.subplots(1, num_models, figsize=(6 * num_models, 5))
    
    # Ensure axes is iterable even if there is only 1 deep learning model
    if num_models == 1:
        axes = [axes]
        
    for ax, (model_name, model_data) in zip(axes, dl_models.items()):
        mdl = model_data['model']
        history = getattr(mdl, 'history', None)
        
        if history and 'train_loss' in history and 'val_loss' in history:
            epochs = range(1, len(history['train_loss']) + 1)
            ax.plot(epochs, history['train_loss'], 'b-', marker='o', label='Training Loss')
            ax.plot(epochs, history['val_loss'], 'r-', marker='o', label='Validation Loss')
            ax.set_title(f'{model_name} Loss Curve', fontsize=12, fontweight='bold')
            ax.set_xlabel('Epochs')
            ax.set_ylabel('Loss')
            ax.legend()
            ax.grid(alpha=0.3)
        else:
            ax.text(0.5, 0.5, 'History not available', ha='center', va='center')
            ax.set_title(model_name)
            
    plt.tight_layout()
    plt.savefig('combined_training_loss_curves.png', dpi=150, bbox_inches='tight')
    print("✅ Training Loss Curves saved as: combined_training_loss_curves.png")
    plt.show()


# =============================================================================
# MAIN EXECUTION
# =============================================================================
if __name__ == "__main__":
    
    print("="*100)
    print("🚀 FAKE NEWS DETECTION - COMBINED DATASET (ISOT + TextDB3 + WELFake)")
    print("="*100)
    
    df_combined = load_and_merge_cleaned_datasets()
    
    print("\n" + "="*100)
    print("📊 DATASET COMPOSITION")
    print("="*100)
    for source, count in df_combined['source'].value_counts().items():
        print(f"    • {source}: {count:,} samples ({count/len(df_combined)*100:.1f}%)")
    
    print("\n" + "="*100)
    print("🔧 INITIALIZING MODELS")
    print("="*100)
    
    models = {}
    print("\n📊 Adding traditional ML models...")
    models['Logistic Regression'] = {
        'type': 'sklearn',
        'vectorizer': TfidfVectorizer(max_features=5000, ngram_range=(1, 2)),
        'model': LogisticRegression(max_iter=1000, random_state=42, n_jobs=-1)
    }
    
    models['Random Forest'] = {
        'type': 'sklearn',
        'vectorizer': TfidfVectorizer(max_features=5000, ngram_range=(1, 2)),
        'model': RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
    }
    
    models['Gradient Boosting'] = {
        'type': 'sklearn',
        'vectorizer': TfidfVectorizer(max_features=5000, ngram_range=(1, 2)),
        'model': GradientBoostingClassifier(n_estimators=100, random_state=42)
    }
    
    models['Naive Bayes'] = {
        'type': 'sklearn',
        'vectorizer': TfidfVectorizer(max_features=5000, ngram_range=(1, 2)),
        'model': MultinomialNB()
    }
    
    if USE_TRANSFORMERS:
        print("\n🔥 Adding transformer models...")
        models['BERT'] = {'type': 'transformer', 'model': TransformerClassifier(model_type='bert', batch_size=16)}
        models['RoBERTa'] = {'type': 'transformer', 'model': TransformerClassifier(model_type='roberta', batch_size=16)}
        models['DistilBERT'] = {'type': 'transformer', 'model': TransformerClassifier(model_type='distilbert', batch_size=16)}
    
    if USE_LSTM:
        print("\n🔤 Adding LSTM model...")
        models['LSTM'] = {'type': 'lstm', 'model': LSTMWrapper(vocab_size=5000, max_len=256, epochs=3, batch_size=64)}
    
    print(f"\n✅ Initialized {len(models)} models")
    
    print("\n" + "="*100)
    print("🎯 TRAINING ALL MODELS ON COMBINED DATASET")
    print("="*100)
    
    results, ensemble_results, trained_models = train_and_evaluate(df_combined, models)
    
    print("\n" + "="*100)
    print("📊 FINAL RESULTS")
    print("="*100)
    
    print("\n📈 Individual Models:")
    print("-" * 80)
    for model_name, metrics in results.items():
        if metrics:
            print(f"\n{model_name}:\n    Accuracy:  {metrics['accuracy']:.4f}\n    Precision: {metrics['precision']:.4f}\n    Recall:    {metrics['recall']:.4f}\n    F1 Score:  {metrics['f1']:.4f}")
    
    if ensemble_results:
        print("\n🤝 Ensemble Models:")
        print("-" * 80)
        for model_name, metrics in ensemble_results.items():
            print(f"\n{model_name}:\n    Accuracy:  {metrics['accuracy']:.4f}\n    Precision: {metrics['precision']:.4f}\n    Recall:    {metrics['recall']:.4f}\n    F1 Score:  {metrics['f1']:.4f}")
    
    all_models = {**results, **ensemble_results}
    valid_models = {k: v for k, v in all_models.items() if v}
    
    if valid_models:
        best_model_name = max(valid_models.items(), key=lambda x: x[1]['f1'])[0]
        best_metrics = valid_models[best_model_name]
        print("\n" + "="*100)
        print("🏆 BEST MODEL")
        print("="*100)
        print(f"\nModel: {best_model_name}\n    Accuracy:  {best_metrics['accuracy']:.4f}\n    Precision: {best_metrics['precision']:.4f}\n    Recall:    {best_metrics['recall']:.4f}\n    F1 Score:  {best_metrics['f1']:.4f}")
    
    print("\n" + "="*100)
    print("📊 GENERATING VISUALIZATIONS")
    print("="*100)
    
    df_results = visualize_results(results, ensemble_results, df_combined)
    create_confusion_matrix(results)
    plot_roc_curves(results)
    plot_training_histories(trained_models)
    
    print("\n" + "="*100)
    print("📝 GENERATING DETAILED REPORT")
    print("="*100)
    
    with open('combined_results_report.txt', 'w') as f:
        f.write("="*100 + "\nFAKE NEWS DETECTION - COMBINED DATASET REPORT\n" + f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n" + "="*100 + "\n\n")
        
        f.write("DATASET INFORMATION:\n")
        f.write(f"    Total samples: {len(df_combined):,}\n")
        f.write(f"    Real news (0): {sum(df_combined['label']==0):,} ({sum(df_combined['label']==0)/len(df_combined)*100:.1f}%)\n")
        f.write(f"    Fake news (1): {sum(df_combined['label']==1):,} ({sum(df_combined['label']==1)/len(df_combined)*100:.1f}%)\n\n")
        
        f.write("    Sources:\n")
        for source, count in df_combined['source'].value_counts().items():
            f.write(f"        • {source}: {count:,} samples ({count/len(df_combined)*100:.1f}%)\n")
        
        f.write("\n" + "="*100 + "\nINDIVIDUAL MODELS\n" + "="*100 + "\n\n")
        for model_name, metrics in results.items():
            if metrics:
                f.write(f"{model_name}:\n    Accuracy:  {metrics['accuracy']:.4f}\n    Precision: {metrics['precision']:.4f}\n    Recall:    {metrics['recall']:.4f}\n    F1 Score:  {metrics['f1']:.4f}\n\n")
        
        if ensemble_results:
            f.write("="*100 + "\nENSEMBLE MODELS\n" + "="*100 + "\n\n")
            for model_name, metrics in ensemble_results.items():
                f.write(f"{model_name}:\n    Accuracy:  {metrics['accuracy']:.4f}\n    Precision: {metrics['precision']:.4f}\n    Recall:    {metrics['recall']:.4f}\n    F1 Score:  {metrics['f1']:.4f}\n\n")
        
        if valid_models:
            f.write("="*100 + "\nBEST MODEL\n" + "="*100 + "\n\n")
            f.write(f"Model: {best_model_name}\n    Accuracy:  {best_metrics['accuracy']:.4f}\n    Precision: {best_metrics['precision']:.4f}\n    Recall:    {best_metrics['recall']:.4f}\n    F1 Score:  {best_metrics['f1']:.4f}\n\n")
        
        f.write("="*100 + "\n")
    
    print("✅ Report saved to: combined_results_report.txt")
    
    print("\n" + "="*100)
    print("✅ TRAINING COMPLETE!")
    print("="*100)
    print("\n📁 Generated Files:")
    print("    • /kaggle/working/combined_dataset.csv - Merged & cleaned dataset")
    print("    • combined_results.png - Comprehensive visualization")
    print("    • combined_confusion_matrix.png - Best model confusion matrix")
    print("    • combined_roc_curves.png - ROC Curves for probabilistic models")
    print("    • combined_training_loss_curves.png - Deep learning training histories")
    print("    • combined_model_results.csv - All results in CSV format")
    print("    • combined_results_report.txt - Detailed text report")
    print("    • checkpoints/ - Training checkpoints")
    
    print("\n" + "="*100)
    print("💾 All results saved in /kaggle/working")
    print("✅ Ready to download from Kaggle Output Files")
    print("="*100)
```
